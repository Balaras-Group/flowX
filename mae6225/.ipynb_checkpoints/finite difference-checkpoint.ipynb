```python
import numpy 
import matplotlib.pyplot as plt
```

```python
Nx = 40
x = numpy.linspace(0.,4*numpy.pi,Nx)
x
```

```python
f = numpy.cos(x)
```

```python
plt.plot(x,f)
```

```python
dfdx_exact = -numpy.sin(x)
plt.plot(x,dfdx_exact)
#il s'agit de la solution exacte
```

```python
dfdx_num = -numpy.zeros(numpy.size(x))
print(dfdx_num)
```

```python
dfdx_num[-1]=dfdx_exact[-1] 
#the last value of the numerical derivative is equal to the last value 
#of the exact solution
dx = x[1] - x[0] # we define dx

dfdx_num[:-1] = (f[1:] - f[:-1])/dx

plt.plot(x,dfdx_exact,'g')
plt.plot(x,dfdx_num,'r')
```
