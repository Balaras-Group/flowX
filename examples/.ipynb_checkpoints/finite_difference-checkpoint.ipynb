```python
import numpy
import matplotlib.pyplot as plt
```

```python
Nx=40
x=numpy.linspace(0.,2*numpy.pi,Nx)
```

```python
f=numpy.cos(x)
```

```python
plt.plot(x,f)
```

```python
dfdx_exact=-numpy.sin(x)
```

```python
plt.plot(x,dfdx_exact)
```

```python
dfdx_num=numpy.zeros(numpy.size(x))
```

```python
print(dfdx_num)
```

```python
#dfdx_num[-1] = dfdx_exact[-1]
```

```python
dx=x[1]-x[0]
#dfdx_num[:-1]=(f[1:]-f[:-1])/dx i can do it this way if i want to use arrays or iteratively
for i in range (Nx-1):
    dfdx_num[i]=(f[i+1]-f[i])/dx
```

```python
plt.plot(x,dfdx_exact,'g')
plt.plot(x,dfdx_num,'r')
```
