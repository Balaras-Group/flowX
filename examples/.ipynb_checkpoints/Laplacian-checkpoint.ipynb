```python
import mae6225
import simulation
mae6225.__version__
```

```python
#Define new parameters
nx, ny = 40 , 40
xmin , xmax = 0.0, 2.0
ymin , ymax = 0.0, 2.0
#Define cell-centered variable names
center_vars=['asol', 'ivar', 'inum', 'error']
```

```python
#Create the grid and data
grid = mae6225.Grid(center_vars,nx,ny,xmin,xmax,ymin,ymax)
#grid is the object, Grid is the class, mae6225 is the library
```

```python
grid.get_values('asol')
```

```python
grid.set_values('asol',1.0)
grid.get_values('asol')
```

```python
simulation.set_function_val(grid,'ivar')
simulation.get_analytical(grid,'asol')
simulation.get_numerical(grid,'ivar','inum')
```

```python
mae6225.io.plot_contour(grid, 'ivar')
mae6225.io.plot_contour(grid,'asol')
mae6225.io.plot_contour(grid,'inum')
```
