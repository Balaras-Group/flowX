```python
import numpy
import matplotlib.pyplot as pyplot

data_crown = numpy.array([773046, 30866364])
data_tshape = numpy.array([706314, 12281904])
data_ref = numpy.array([7e5, 4e7])
```

```python
pyplot.rc('font', family='serif', size=16)
pyplot.figure()
pyplot.xlabel('Ann Search        Classical Search')
pyplot.ylabel('log C')
pyplot.plot(numpy.log(data_ref), numpy.log(data_crown))
pyplot.plot(numpy.log(data_ref), numpy.log(data_tshape))
pyplot.legend(['CROWN', 'T'])
pyplot.axis('scaled', adjustable='box')
pyplot.axis('equal')
pyplot.ylim(12, 18)
pyplot.show()
```
